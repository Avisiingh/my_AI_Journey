##### Research Question 1

- Goal: Find which state has the most expensive housing market.
- How: Group the dataset by the state column, then calculate the mean house price for each state.
- Why: This lets us compare “typical” prices by state and identify the most expensive ones.
- Next: Visualize the results (bar chart or boxplot) to see price differences clearly.

```python
mean_price_by_state =df.groupby("state")["price_usd"].mean().sort_values(ascending=False)
mean_price_by_state.head()
```

- Make a bar chart using mean_price_by_state.
- Sort states from highest to lowest mean price.
- Label x-axis State.
- Label y-axis Mean Price [USD].
- Title the chart Mean House Price by State.

```python
## creating bar chart
mean_price_by_state.plot(
    kind="bar",
    xlabel="state",
    ylabel="price [USD]",
    title="mean_price_by_state"
);
```

Create a new column in df called "price_per_m2". This should be the price for each house divided by it's size.

```python
df["price_per_m2"] =df["price_usd"] /df["area_m2"]
df.head()
```

Task:

- Use groupby on the state column and calculate the mean of price_per_m2 for each state.
- Sort the results from highest to lowest mean.
- Plot a bar chart of these values.
- Label the x-axis State.
- Label the y-axis Mean Price per m² [USD].
- Title the chart Mean House Price per m² by State.

```python
(
    df
    .groupby("state")["price_per_m2"]
    .mean()
    .sort_values(ascending=False)
    .plot(
        kind="bar",
        xlabel="state",
        ylabel="mean price per m2[USD]",
        title ="Mean House Price per M^2 by State"
    )
);
```

##### Research Question 2

Q - Is there a relationship between home size and price?
  -Goal: Check if home size affects price.
  - Why: We already saw location matters. Now we want to see if bigger homes cost more.
  - How: Make a scatter plot with
     -   x-axis = area_m2 (home size)
     -   y-axis = price (home price)
  - This lets you visually spot if prices go up as home size increases (a positive correlation).

```python
plt.scatter(x=df["area_m2"],y=df["price_usd"])
plt.xlabel("area [sq meters]")
plt.ylabel("price [USD]")
plt.title("price vs Area");
```

 Using the corr method, calculate the Pearson correlation coefficient for "area_m2" and "price_usd".

```python
p_correlation =df["area_m2"].corr(df["price_usd"])
print(p_correlation)
```

The correlation coefficient is over 0.5, so there's a moderate relationship house size and price in Mexico. But does this relationship hold true in every state? Let's look at a couple of states, starting with Morelos.

```python
df_morelos =df[df["state"]=="Morelos"]
df_morelos.head()
```

```python
plt.scatter(x=df_morelos["area_m2"],y=df_morelos["price_usd"])
plt.xlabel
plt.ylabel
plt.title()
```
